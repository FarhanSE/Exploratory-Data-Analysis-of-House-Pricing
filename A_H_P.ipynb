### Here it does not show all the columns so let's try in the next cell

```python
# To check missing values of all columns
for i in train.columns:
    print(f"{i}: " + str(train[f'{i}'].isnull().sum()))
```

### there are many columns where we have huge NA values

```python
# Here we made a list of feature which has NA values
feature_na_values = [ feature for feature in train.columns if train[feature].isnull().sum()>1]
```

### Here we will find the relation between missing values and sales price

```python
for feature in feature_na_values:
    data = train.copy()
    # let's make a veriable it indicates 1 if values are missing or 0 otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### Here the Relation of Missing Values and Dependent feature clearly visible. So we have to fill these missing values with something meaningful.

ID feature is not required so we will drop it

### Numerical Values

```python
numerical_values = [feature for feature in train.columns if train[feature].dtype != 'O']
print("Number of numerical columns:", len(numerical_values))
```

```python
train[numerical_values].head()
```

### List of columns that contains year information

```python
list_year_feature = [feature for feature in train.columns if 'Yr' in feature or 'Year' in feature]
list_year_feature
```

```python
for feature in list_year_feature:
    print(feature, train[feature].unique())
```

### Let's analyze is there any relation between year and sale price

```python
train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Sale Price')
plt.title('Year Sold VS Sale Price')
plt.show()
```

```python
for feature in list_year_feature:
    if feature != 'YrSold':
        data = train.copy()
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.show()
    
```

```python
discrete_feature = [feature for feature in numerical_values if len(train[feature].unique())<25 and feature not in list_year_feature+['Id']]
print("Number of discrete feature:", len(discrete_feature))
```

```python
train[discrete_feature].head()
```

### Let's check relation between discrete veriables and SalePrice

```python
for feature in discrete_feature:
    data = train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("Sale Price")
    plt.title(f"Sale Price VS {feature}")
    plt.show()
```
